```python
class Memberinfo:
    def __init__(self):
        import pandas as pd
        import numpy as np
        import matplotlib.pyplot as plt
      
        #memberinfo = pd.read_csv('성도정보.csv', encoding='ANSI' )
        self.__memberinfo = pd.read_csv('성도정보.csv' )
        # 데이터 프레임 생성

    def save_memberinfo(self):
        self.__memberinfo.iloc[:,1:].to_csv('성도정보.csv')

    def add_memberinfo(self,newline_memberinfo):
        # 열 개수가 맞는지 확인
        #print(newline_memberinfo)
        self.__memberinfo = self.__memberinfo.append(newline_memberinfo, ignore_index=True)
        #print(memberinfo)




    def see_preview_memberinfo(self):
        print(self.__memberinfo)

    def see_memberinfo_rowindex(self, startrow, endrow):
    # 0부터 len()까지
        for i in range(startrow, endrow+1):
            print ( self.__memberinfo.iloc[i] , '\n\n' )

    def see_memberinfo_iloc(self, iloc_row, iloc_column):
        # at.iloc[row1,column1]
        print ( self.__memberinfo.iloc[ iloc_row, iloc_column ] )


    def see_columns(self):
        print(self.__memberinfo.columns)
        print (self.__memberinfo.index.name)

###################################################################
class Attendinfo:
    def __init__(self):
        import pandas as pd
        import numpy as np
        import matplotlib.pyplot as plt

        #attendinfo = pd.read_csv('출석정보.csv', encoding='ANSI' )
        self.__attendinfo = pd.read_csv('출석정보.csv' )
        # 데이터 프레임 생성


    def save_attendinfo(self):
        self.__attendinfo.iloc[:,1:].to_csv('출석정보.csv')

    def add_attendinfo(self, newline_attendinfo):
    # 열 개수가 맞는지 확인
        self.__attendinfo = self.__attendinfo.append(newline_attendinfo, ignore_index=True)
    



    def see_preview_attendinfo(self):
        print(self.__attendinfo)


    def see_attendinfo_rowindex(self, startrow, endrow):
    # 0부터 len()까지
        for i in range(startrow, endrow+1):
            print ( self.__attendinfo.iloc[i] , '\n\n' )

    def see_attendinfo_iloc(self, iloc_row, iloc_column):
        # at.iloc[row1,column1]
        print ( self.__attendinfo.iloc[ iloc_row, iloc_column ] )


    def see_columns(self):
        print(self.__attendinfo.columns)
        print(self.__attendinfo.index.name)

##################################################################
class Prepare:
    def make_iloc_rowindex(self):
        print('추출할 행번호를 입력하세요.')
        print('범위는 - 을 이용하여  1-20과 같이 지정하며,')
        print('각 원소는 , 로 구분하여 25-80,91,92,93 와 같이 입력합니다.')
        print('또한 모든 띄어쓰기는 제거됩니다. ex) 10 4 는 104로 입력됩니다.')
        
        self.temprow = input(' >>>>>>>>> ')
        self.temprow = self.temprow.replace(" ",'')
        self.temprow = self.temprow.split(',')
        print('입력하신 값은 다음과 같습니다.')
        print (self.temprow )

        self.rowindex = []
        # row1.extend(list(range(5,30)))
        for i in self.temprow:
            if '-' in i:
                self.temp = i.split('-')
                self.rowindex.extend(list(range(int(self.temp[0]),int(self.temp[1])+1)))
            else:
                self.rowindex.append(int(i))

        self.rowindex = list(set(self.rowindex))
        print(self.rowindex)
        return self.rowindex.sort()

    ###3################################################333
class UesrInterface(Prepare, Attendinfo, Memberinfo):
    def __init__(self):
        print(' 디모데 목장 프로그램을 이용해 주셔서 감사합니다. \n 무엇을 도와드릴까요?')
        print(' 성도리스트를 열람하시려면 1을, 출석정보를 확인하시려면 2를 입력해 주세요.')
        self.select = input('>>>>>>>>>>>')
        self.start()



    def administor(self):
        print('무엇을 원하시나요?')


    def start(self):

        if self.select == '1':
            see_preview_memberinfo()

        elif self.select == '2':
            see_preview_attendinfo()

        elif self.select == '0':
            return 

        elif self.select == 'password':
            self.administor()
        else :
            print('잘못 입력하셨습니다. 다시입력해 주세요.\n 종료를 원하시면 0번을 입력해 주세요')
            self.select = input('>>>>>>>>>>>')
            self.start()







##################
################
###########
#######
##

test1 = UesrInterface()

```
